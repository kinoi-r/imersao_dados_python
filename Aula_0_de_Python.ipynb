Como a Colab é linkada com a Gemini, a inteligência artificial da Google, podemos acessar Generate with AI e solicitar que traduza o nome das colunas pra português. Não é necessário copiar o nome das colunas, basta dizer "traduza para mim o nome das colunas dataframe df" que já é suficiente. Ela vai gerar um código pra conferirmos, se estiver tudo certo, basta rodar e ele renomeará as colunas com os nomes em português.


DICA: Em um arquivo REAL, devemos realizar renomeações unicamente em novos arquivos, para não perdermos os arquivos originais.

```python
df.rename(columns={
    'ano': 'ano',
    'nivel_experiencia': 'senioridade',
    'tipo_emprego': 'contrato',
    'titulo_cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'proporcao_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}, inplace=True)


print("Novos nomes das colunas:")
print(df.columns)
```

```python
df['senioridade'].value_counts()
```

Agora utilizamos df['nivel_experiencia'] com o método value_counts() pra chamar o arquivo df e revelar a frequência, contar os valores de cada categoria SE (profissional sênior), MI (profissional mid, intermediário), EN (profissional junior, iniciante) e EX (executivo, liderança) dentro da coluna de nível de experiência.

```python
df['contrato'].value_counts()
```

Podemos repetir com outras colunas. Nesse resultado, vemos que na categoria FT (Full time, integral), há 132563 funcionários. CT (Contrato temporário), há 394 funcionários. PT (Part time, meio período), há 376 funcionários. FL (freelance) há 16 funcionários.

```python
df['remoto'].value_counts()
```

Repetindo com "remoto" pra ter melhor visualização dos contratos remotos da empresa. Em "0" indica funcionários presenciais, em "100" indica funcionários remotos e em "50" indica vagas híbridas.

```python
df['tamanho_empresa'].value_counts()
```

Aqui "M" indica empresas médias, que são em maior número, de acordo com o resultado. "L" indica empresas grandes e "S" empresas pequenas.

Note como há uma grande quantidade de siglas aparecendo. O ideal é deixar tudo com nomes intuitivos e facéis de entender, especialmente se houver uma apresentação desse trabalho há alguém interessado. Portanto, podemos usar novamente a função Generate with AI pra renomear o nome das categorias das colunas apresentadas.

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

print("Novas categorias na coluna 'senioridade':")
print(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Integral',
    'CT': 'Temporário',
    'PT': 'Meio Período',
    'FL': 'Freelance'
})

print("Novas categorias na coluna 'contrato':")
print(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

print("Novas categorias na coluna 'tamanho_empresa':")
print(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})

print("Novas categorias na coluna 'remoto':")
print(df['remoto'].value_counts())
```
